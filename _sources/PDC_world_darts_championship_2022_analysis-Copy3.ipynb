# 2. About the data

The data I scraped is divided into one row for every dart thrown and some extra rows for score change events. the dataset has the following fautures:

**Type:** what type of event (dart thrown or score change)\
**points:** how many points scored on that darts (e.g. 60 when a player throws a triple 20)\
**segment:** a dartsboard is divided into multiple segments (single, double, triple, bull, bullseye). Every segment has a corresponding number\
**event_points:** in what number was thrown (e.g. 20 when a player throws a triple 20)\
**double_attempt:** In darts you have to end a leg by throwing in a double. this shows how many times the player tried to throw in double\
**team:** home or away\
**leg_points_home/leg_points_away:** In darts a must go from 501 points to 0 points to win a leg. If a player wins 3 legs, he wins a set. When a player wins a predefined amount of sets, he wins the game. This feature shows the players leg points for each throw (e.g. 345). \
**leg_score_home/leg_score_away:** Same as above but then the leg score. (e.g. 2 when the player has won 2 legs)
**set_score_home/set_score_away:** Same as above but then the set score. (e.g. 3 when the player has won 3 sets)
